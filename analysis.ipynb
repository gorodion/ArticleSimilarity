```python
import pandas as pd
import numpy as np
from sklearn.metrics.pairwise import cosine_similarity
```

```python
df = pd.read_csv('data/elec_base.csv')
df = df.drop_duplicates(ignore_index=True)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer as Vect
vect = Vect(min_df=5)
X = vect.fit_transform(df.content.fillna(''))
```

```python
cosine = cosine_similarity(X)
```

```python
cosine = np.triu(cosine, 1)
```

```python
spam = pd.DataFrame({'Заголовок 1': df.title.values, 
                     'Заголовок 2': df.title.loc[cosine.argmax(1)].values, 
                     'Мера близости': cosine.max(1)})
```

```python
spam = spam.sort_values('Мера близости')[-55::-1]
```
